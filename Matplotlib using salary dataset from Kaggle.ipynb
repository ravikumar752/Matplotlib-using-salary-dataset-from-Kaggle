```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("D:\python books\data sets\Salary_data.csv")  # load dataset from your local computer
X = data["YearsExperience"]  
Y = data["Salary"]
plt.plot(X,Y)
points = plt.scatter(X,Y)
points.set_facecolor("red")
plt.title("Scatter Plot")
plt.xlabel("Working Experience (years)")
plt.ylabel("Annual Wage (dollars)")
plt.show()
```
